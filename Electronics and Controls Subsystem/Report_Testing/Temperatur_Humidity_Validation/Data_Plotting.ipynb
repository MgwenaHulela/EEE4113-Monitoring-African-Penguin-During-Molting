```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("observatory_weather_data.csv", parse_dates=['Timestamp'])

# Set the timestamp as index for better plotting
df.set_index('Timestamp', inplace=True)

# Create subplots: Temperature and Humidity side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# Temperature plot
df[['AHT20+BMP280 Temperature', 'BME280 Temperature (Reference)']].plot(ax=ax1)
ax1.set_title("Temperature Readings")
ax1.set_ylabel("Temperature (°C)")
ax1.set_xlabel("Time")
ax1.grid(True)

# Humidity plot
df[['AHT20+BMP280 Humidity', 'BME280 Humidity (Reference)']].plot(ax=ax2)
ax2.set_title("Humidity Readings")
ax2.set_ylabel("Humidity (%RH)")
ax2.set_xlabel("Time")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
